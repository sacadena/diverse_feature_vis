```python
%matplotlib inline
import numpy as np
import os
import matplotlib.pyplot as plt
import diverse_vis_functions as dvis
```

### Diverse templates with no natural image prior

```python
layer_name = 'conv3/conv3_1'
distance_layer_name = 'conv2/conv2_2' # set to None for pixel space distance
unit_num = 21 # Feature map
num_images = 6
dist_weight = 0.01 # referred to \lambda in the paper
naturalness = 0.0
```

```python
subspace_net = dvis.SubspaceNet(layer_name, unit_num, num_images, dist_weight, naturalness, feature_layer_name=distance_layer_name)
images, predictions, min_distance, mean_distance, loss = subspace_net.map_subspace(callback=dvis.callback, callback_every=500, max_iter=2000)
avg_rate = predictions.mean()
print('Done after {:d} iterations (avg rate = {:.2f})'.format(len(loss), avg_rate))
```
